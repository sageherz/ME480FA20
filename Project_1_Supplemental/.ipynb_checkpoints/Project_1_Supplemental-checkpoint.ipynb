## Workstation Setup

* Configure your M480 workstation as shown in the figure by removing the load inertial module and installing the gate module. The gate limit switches plug into the 8 pin connector (Gate 1).
* The remote sensor is located in the bag of accessories supplied with the rig. The remote sensor plugs into the 4 pin Emitter/Detector plug located on the top right corner of the rig. 
* The motor should be connected to Motor 2 and the gate limit switches and the gate should be installed.  The remote sensor can be positioned up to about 20 inches from the gate. 

## Gate FSM Operation

* You can download the code from jupyter by opening the [ME480L_Project1_Gate file](ME480L_Project1_Gate.ino)
* The program starts with the gate down in the WAIT state. If you hear the motor trying to run, ensure the gate is closing the lower limit switch. Typically this only happens when you first start the program.  (If you find you need more torque to close the switch, you can increase the voltage to the motor applied when it tries to lower the gate by increasing the  ```motorDownPWMCounts``` variable in the code). 
* The green ready LED should be lit. In this state you may change the program between near and far runs by pressing BTN 3 or BNT 4. The mode will be shown in the lower line of the LCD display. 
* If the sensor is blocked you will see the blue run light come on. This will happen regardless of the STATE.
* Position your Zumo so it is pointed at the gate and blocking the remote sensor. Roll it backward **just** until the blue light goes off, indicating that the Zumo is no longer blocking the sensor.
* Pressing BTN 1 will move the system into armed state. The green led will change to a red blinking led. At this point you are ready to run your Zumo program.
* Once the Zumo's presences activates the detector, the blue light will momentarily light and the blinking red light will change to a steady red light. After a specified time interval the gate will open.  It will remain open for a second specified time interval and then close. 


During the demonstration we will provide the gate timing.  You can practice with the gates, by entering your own timing by adjusting the timer durations found at the top of the code and described in the comments:

``` c
//+++++++++++ Timer variables ++++++++++
unsigned long timer1DurationFar = 3000;  //time(ms) from start when gate will open (FAR)
unsigned long timer2DurationFar = 2000;  //time(ms) gate will remain open (FAR)
unsigned long timer1DurationNear = 2000; //time(ms) from start the gate will open (NEAR) 
unsigned long timer2DurationNear = 800;  //time(ms) gate will remain open (NEAR)
//++++++++++++++++++++++++++++++++++++++
```

## Evaluation

Your challenge is to ensure the zumo arrives on the elevator platform before it moves (as indicated by the gates lowering) and not fall off!  We'll evaluate how well the zumo's open-loop controller worked by using the zones shown below.  

Success is ensured by the zumo being entirely in the safe zone (the size of the elevator).  Arriving on the target (Centered) does recieve a bit of a bonus, but is not required.  There are increasing penalties for being in the Dangling, or 50/50 zone.  The lowest points are awarded for any part of the zumo being outside the 50/50 zone.

**Your zumo will be scored according to which zone it is *entirely* within**.

If you like, you can download and print an 8.5x11 target sheet that has 3 of the zones in order to quickly evaluate your postion [target](Project_1_target.pdf), but this is not required.

## Demonstration Day Procedure

* At the start of Lab class, have the gate-control-code loaded on your workstation and have some kind of scale, ruler, tape measure, [or printed scale](https://www.printablerulers.net/) handy.  
* We will provide you the gate timing parameters for that day's demonstration at the beginning of class
* You will have time to setup your zumo for each condition and then demonstrate


* Before you run, you will be asked to specifically demonstrate that your state machine meets the design specifications. 
* You will run the zumo for 2 different distances.  Although your modeling should predict the change in perforance with different payloads, you will only demonstrate you zumo **unloaded**.

* Each run will be scored on distance accuracy and timing. 
* You will submit the version of the code used for testing DURING the lab period.
* You will also be asked to briefly explain some of your modeling choices and programming choices in class

* To capture your distance accuracy, we we'll ask you to make your best estimate of the zone the zumo achieved and record the postition with a quick pic of zumo with a target (a piece of tape with an x, a little piece of a post-it, a mark on the surface, etc) and a scale like the image shown below.

## Project 1 Report

By the beginning of the next Lab class, you will need to submit your report in the form of the Project_1_Report jupyter notebook.  The condensed report format is explained below.  Ase with other jupyter notebooks, you can add cells as needed, but **DO NOT SPLIT OR DELETE CELLS**.

Hand drawn figures and equations are acceptable, but where a bit of narration is required, type it in a markdown cell.

### Finite State Machine

Include your state transition diagram and table

### Model Development and Validation

This can include hand written notes, but make assumptions clear (add in text if needed) and include the final validated model.

### Performance Assement

Include the images of your zumo's performance and make a brief quantative comparison of the actual performance and the predicted performance in terms of timing and distance along with your assessment.

### Next Steps

If your zumo did not perform as expected, you can address issues after the demonstration and before the report is due and report on that effort in this section.  However, this Next Steps section should only include a discussion of issues that were both identified **and** an attempt was made to address them.  For exmaple, if your performance on demonstration day was poor and you later realize a code mistake, you can explain the mistake here and show updated results of the performance day runs as evidence the performance improved.  Points received for Next Steps will be used to improve your performance grade.    


**Upload your matlab, arduino, and data files to the jupyter folder so they are included in your submission**
