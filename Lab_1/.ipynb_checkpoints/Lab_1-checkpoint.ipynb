# *Model Development:* DC Motor

A direct current motor uses windings of wire to generate a magnetic field which causes the *rotor* to turn when placed in a magnetic field.  As the rotor aligns with the external magnetic field, the *commutator* switches the current to a subsequent coil of wire which causes it to again rotate.  The process is repeated for every segment of the commutator causing the rotor to turn continuously.  It has been found experimentally that the torque is proportional to the current and the angular velocity is proportional to the voltage (something you discussed in Systems).

# *Controller Design:* DC Motor Speed and Direction

We know from observation of DC motors that the speed is related to the voltage it receives, but we can only supply a fixed 5V to our motor with our workstation.  This is not an uncommon problem.  We'll need a way to somehow change the voltage the motor "sees" even though our voltage supply remains fixed.  Quite a conundrum!

## *Model Development* Pulse Width Modulation 

Although changing the DC voltage supplied to a DC motor will indeed change the motor's speed, digital circuits (like the Arduino board) typically operate with a fixed voltage.  Digital control is not suited to infinitely vary a voltage, but it **CAN** modulate a signal on or off.  Therefore, a different method of controlling the Zumo motor speed is required if we are going to control it with a digital microcontroller (a computer whose dedicated purpose is the control of equipment). 

Pulse width modulation will be used to achieve this where the frequency of a signal is CONSTANT, and just the DUTY CYCLE (the % time the signal is on vs off over one period) is varied.  The figure below shows that by 
changing the duty cycle (and NOT the fequency) the average voltage of a signal changes.  




If a system has enough capacitance (from a system modeling perspective, so this can be related to electrical capacitance or mass for example), it will naturally integrate (filter, smooth) a high frequency signal.  That is to say, if the pulse width modulated frequency is chosen to be high enough above the time constant of a system, it will "see" the PWM signal as a constant DC signal. ## *Model Development:* Transistor Model

#  *Controller Design:* H-Bridge Circuit

Although we did not need much amplfiication of the current in the exmaples above, the transistor can amplify the base current and create higher currents through the collector and emitter.  Although a single transistor could allow us to amplify the PWM signal to control the speed of the motor, your motor will need to change direction (for the gate and for the Zumo) and so you also need a circuit that can do this *WITHOUT* requiring you to manually disconnect and reversing the wires.  

The ***H-Bridge*** circuit can change the direction of a motor, without connecting/disconnecting wires, by using 4 transistors arranged in an "H" shape.  Because transistors can act as an open a circuit when no current is provided to the base, they can be thought of as a kind of switch.  By alternating the switches that are closed and open, the direction of the current passing though the motor can be changed and thus the direction of the motor can be changed as shown below.

# *Model Development:* Switches

To achieve the design goals for the gate's operation, you will need to know when the gate is up and when it is down, so in a way, we need to include a model of how switches work.  You will use the limit switches installed in the gate base that will close when the gate presses against them.  You will also use momentary button switches as a user input in order to command the gate when to open and close.

For both of these components, the electrical operation is a simple switch that closes a circuit.  On the ME480 workstation, the gate limit switches and the user input buttons are configured with one side connected to ground and the other connected to an Arduino digital pin. For the gate's "up" or "high" limit switch, this corresponds to the circuit below:



In this case, the arduino pin must be "nominally" at 5V so that when the switch is closed, pin 10 should read 0V.  When the switch is opened, pin 10 should read 0V.  Using a "pullup resistor," this can be accomplished without causing a short circuit. Pullup resistors are described in the video below.

````python
%%html
<iframe width="900" height="540" src="https://www.youtube.com/embed/wxjerCHCEMg?start=124&end=187&version=3" frameborder="0" allow="accelerometer; autoplay; encrypted-media; gyroscope; picture-in-picture" allowfullscreen></iframe>
````

Luckily, the Arduino has the ability to configure a digital pin using an **internal** pull-up resistor, so no change to the circuit is required! Configuring the pins this way is as simple as one line of code.

Use the code below to give this a try.  We'll use the serial monitor to indicate if the switch is closed or if the switch is open.

```c
const int upLimitSwitch  = 10;   

void setup() {
  pinMode(upLimitSwitch, INPUT_PULLUP); // initialize the pin connected to the "up" limit switch as an input
  Serial.begin(9600); //initialize serial communication
}

void loop() {
  int sensorVal = digitalRead(upLimitSwitch); //read the current value of the pin connected to the limit switch 
  //sensorVal is declared as an integer
  //in this case, it can be 1 (if the pin is at 5V) or 0 (if the pin is connected to ground)  
  
  //print sensorVal to the serial monitor so you can see it.
  Serial.println(sensorVal)    
}
```

  digitalWrite(37,HIGH);
  int S_BTN = digitalRead(BTN);
  int S_UP = digitalRead(UP);
  int S_DOWN = digitalRead(DOWN);
  int S_ESTOP = digitalRead(ESTOP);

  //block 2
  V4 = V3&&!S_UP&&!S_BTN&&S_ESTOP;
  V5 = V2&&!S_DOWN;
  V6 = V2&&S_DOWN;
  V7 = V3&&S_BTN||V3&&!S_ESTOP;
  V8 = V3&&!S_DOWN&&!S_BTN&&S_ESTOP||V3&&!S_BTN&&S_UP&&S_DOWN&&S_ESTOP;
  V9 = V1&&!S_UP;
  V10 = V1&&S_UP;

  //block 3
  V1 = V8||V10;
  V2 = V4||V6;
  V3 = V5||V7||V9;

  //block 4
  if (V1 == 1){
    Serial.println("Gate up");
    analogWrite(46,100);
    digitalWrite(44,LOW);
  }
  if (V2 == 1){
    Serial.println("Gate down");
    analogWrite(44,50);
    digitalWrite(46,LOW);
  }
  if (V3 == 1){
    Serial.println("Wait");
    digitalWrite(44,LOW);
    digitalWrite(46,LOW);
  }
}
