# Project 3: Robocross

You've done a lot this semester with your robot. You've modeled its behavior and controlled its distance traveled using open-loop control and a finite state machine. You've designed a "dead reckoning" waypoint navigation system using a combination of several feedback controllers and a finite state machine. Now, it's time to add one more piece of functionality. It is our hope that you come out of the semester with a nearly autonomous robot you can be proud of, so we'd like you to add the ability to follow a line to your robot's growing repertoire of antics.

For project 3, your robot will race around an autocross-style track. Your score will be based on the time it takes your robot to complete the course, with time penalties for large deviations (over about 2cm from track center). Some basic properties of the task are listed below:
* The track will be a black 1.9 cm (0.75 in) wide black stripe on white paper.
* The track will consist of straight and constant-radius segments.
* There will be "steps" in the track's lateral position on a straight section.
* There will be an "obstacle" on one of the longer straights, which your robot will need to navigate around using encoders and the gyro as in project 2. The obstacle will be represented by your Zumo's box.

An example track is shown in the figure below. Your run will be electronically timed, and your score will be computed as the total time taken to reach the finish line divided by the track distance. If you do not make it all the way to the finish line, your score will be computed based on the distance you were able to go and the approximate total time elapsed at the time your robot leaves the track for the last time. 

There are pdfs available on moodle of "track elements" that can be printed on 8.5x11 paper to test straight track, steps, and turns.  The full track will be provided in advance of your in-lab zumo-prep day (last week of class).

If you don't have a space that meets the requirements, we can reimburse you for the purchase of backboard such as a piece of readily available [hardboard](https://www.homedepot.com/p/1-8-in-x-2-ft-x-4-ft-Tempered-Hardboard-Actual-0-115-in-x-23-75-in-x-47-75-in-109112/202585358), a piece of [corregated plastic](https://www.homedepot.com/p/Coroplast-24-in-x-36-in-Twin-wall-Plastic-Sheet-COR-2436/202090128), or piece of [white Foam Board](https://www.staples.com/crescent-heat-activated-fome-cor-board-3-16-x-24-x-36-white-11105-2436c/product_24412841).

# Appendix: Zumo Line Sensor Configuration

In order to use a thin black line for our track, we will need you to configure your zumo to disable the right and left proximity sensors, and enable the right-middle and left-middle line sensors. You can do this by moving a jumper on the Zumo's front sensor board.

To remove the front sensor board (located just behind the blade on the bottom of the zumo), you only need to **gently** pull the board from the socket, straight down. **Make sure your zumo is off and unplugged from your computer** before attempting this. Once you have removed the front sensor board, move the blue jumpers so they are connecting pins 1 and 2 as shown below.

This will ensure that on the zumo's right side, the jumper connects pin 4 with DN4. From the factory, the jumper is positioned so that pin 4 is connected to RGT (the right proximity sensor).  But since we need the line sensor, moving the jumper will disable the right proximity sensor and ENABLE the right side line sensor.

On the zumo's left side, the jumper should connect pin 20 with DN2, enabling the left line sensor and disabling the left proximity sensor (LFT). 

Once you have configured your sensor board so that pin 20 is connected to DN2 and pin 4 is connected to DN4, you can reinstall your front sensor board. **be very careful to ensure that each pin is aligned with the correct hole during installation**. To prevent an accidental short circuit that could damage your robot during reinstallation, do not plug your zumo in and turn it on until after the board has been reinstalled.

With your line sensors configured, you are ready to get started!

# Appendix: Starter Code

``` javascript

#include <Wire.h>
#include <Zumo32U4.h>

Zumo32U4LineSensors lineSensors;
Zumo32U4Motors motors;
Zumo32U4ButtonA buttonA;
Zumo32U4LCD lcd;
Zumo32U4ProximitySensors proxSensors;

//number of sensors on the Zumo.
//the robot has 5 sensors, but we will only use middle right and middle left.
#define NUM_SENSORS 5
//array of sensor values
unsigned int lineSensorValues[NUM_SENSORS] = {0,0,0,0,0};

//a variable to hold the integer value from the distance sensor.
//this variable goes from about 2 to about 6. 6 is very close, 2 is far.
int proxDist = 0;

//routine for calibrating the sensors
//this function was written by Ben Schmidel from Pololu, copied here.
//this allows you to calibrate the sensors right when zumo powers up.
//The calibration sets the threshold value for the sensors to distinguish 
//between the track (dark) and surroundings (light).
//INSTRUCTIONS:  During calibration mode, slide the zumo laterally across 
//the track so that all of the line sensors are alternately exposed to a 
//a light surface and a dark surface.  
void calibrateLineSensors()
{
  // To indicate we are in calibration mode, turn on the yellow LED
  // and print "Line cal" on the LCD.
  ledYellow(1);
  lcd.clear();
  lcd.print(F("Line cal"));

  for (uint16_t i = 0; i < 400; i++)
  {
    lcd.gotoXY(0, 1);
    lcd.print(i);
    lineSensors.calibrate();
  }

  ledYellow(0);
  lcd.clear();
}


void setup()
{
  Serial.begin(115200);
  //initialize 5 line sensors... middle, and inside left, inside right.
  //NOTE your jumpers on the underside of the zumo's front sensor board must
  //be moved so that they connect DN4 and DN3. This will de-activate the left and
  //right proximity sensors.
  lineSensors.initFiveSensors();
  //initialize only the front proximity sensor
  proxSensors.initFrontSensor();

  // Wait for button A to be pressed and released.
  lcd.clear();
  lcd.print(F("Press A"));
  lcd.gotoXY(0, 1);
  lcd.print(F("to calib"));
  buttonA.waitForButton();

  //this will loop until you move the zumo across the line several times by hand.
  calibrateLineSensors();
}

void loop()
{
  //update line sensor array reading
  lineSensors.readCalibrated(lineSensorValues);
  //read proxSensor to see if there's an obstacle.
  proxSensors.read();

  // Our "error" is how far we are away from the center of the
  // line. This is the integer version in counts
  int error = readLineAAB(lineSensorValues);
  //Below is an approximate conversion from counts to meters. We have
  //a useable detection range of about 1 cm in each direction from the line.
  //Your controller must be pretty good to avoid losing the line!
  float lineerror_m = error*0.0127/1000;//this scaling is approximate and
  //YOU MAY WANT TO ADJUST THIS SCALING.
    
  //this extracts a value from the data collected from the 
  //proximity sensor read above.
  int proxCounts = proxSensors.countsFrontWithRightLeds();
  //print the error from the line and the counts for the forward proximity sensor
  Serial.print(lineerror_m,4);
  Serial.print("\t");
  Serial.println(proxCounts);
  Serial.println();

  //short delay to allow the analog to digital converters to settle.
  delayMicroseconds(1000);
}


//prof. Brown's simple function to read a .75" line and turn into an integer "error"
int readLineAAB(unsigned int vals[5]){
  //function returns integer counts representing error from line center
  //values 1, 2, 3 are the actual 3 sensors wea are using
  int pos = vals[1]-vals[3];
  return pos;
}
```
